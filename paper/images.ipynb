```python
import kerrgeopy as kg
from math import pi, cos
import matplotlib.pyplot as plt

orbit = kg.StableOrbit(0.998,3,0.6,cos(pi/4))
fig, ax = orbit.plot(0,15, lw=1.5, tau=7,axes=False)
plt.tight_layout()
plt.savefig('generic.png',dpi=300)
```

```python
orbit = kg.StableOrbit(0.998,3,0.5,1)
fig, ax = orbit.plot(0,30, lw=1, tau=10,axes=False)
plt.tight_layout()
plt.savefig('equatorial.png',dpi=300)
```

```python
orbit = kg.StableOrbit(0.998,3,0,cos(pi/8))
fig, ax = orbit.plot(0,30, lw=1.5, tau=10,axes=False)
plt.tight_layout()
plt.savefig('spherical.png',dpi=300)
```
