## Separatrix

```python
def sub_zeros(txt):
    txt1 = sub("0\.0*,","0,",txt)
    txt1 = sub("0\.0*]","0]",txt1)
    txt1 = sub("1\.0*,","1,",txt1)
    txt1 = sub("1\.0*]","1]",txt1)
    return txt1
```

```python
num_values = int(1e4)
sep_values = np.zeros((num_values,3))
i = 0
while i < num_values:
    parameters = np.round([np.random.rand(), np.random.rand(), np.random.rand()*2-1], \
                          decimals = 3)
    if parameters[0] != 1:
        sep_values[i] = parameters
        i += 1
sep_values = pd.DataFrame(sep_values,columns=["a","e","x"])
mathematica_sep_output = sep_values.apply(lambda x: \
                                        session.evaluate(wlexpr(
                                        sub_zeros("KerrGeoSeparatrix[{:},{:},{:}]".format(*x)))) \
                                        ,axis=1)
python_sep_output = sep_values.apply(lambda x: separatrix(*x),axis=1)
```

```python
plt.hist(python_sep_output,bins=50)
plt.xlabel("Separatrix")
```

```python
sep_differences = pd.Series(abs(python_sep_output-mathematica_sep_output)/python_sep_output)
sep_differences.describe()
```

```python
parameters = ["a","e","x"]
fig, ax = plt.subplots(1,3,figsize=(15,4))
for i in range(3):
        ax[i].scatter(sep_values[sep_values.columns[i]],np.log10(sep_differences),s=1)
        ax[i].set_xlabel(parameters[i])
        ax[i].set_ylabel("separatrix difference")
```

## Elliptic Functions

```python
def elliptic_parameters(a,p,e,x):
    r1,r2,r3,r4 = radial_roots(a,p,e,x)
    r_plus = 1+np.sqrt(1-a**2)
    r_minus = 1-np.sqrt(1-a**2)
    
    h_r = (r1-r2)/(r1-r3)
    h_plus = (r1-r2)*(r3-r_plus)/((r1-r3)*(r2-r_plus))
    h_minus = (r1-r2)*(r3-r_minus)/((r1-r3)*(r2-r_minus))
    k_r = np.sqrt((r1-r2)*(r3-r4)/((r1-r3)*(r2-r4)))
    
    return h_plus, h_minus, h_r, k_r
```

```python
elliptic_output = np.apply_along_axis(lambda x: elliptic_parameters(*x),1,values)
elliptic_df = pd.DataFrame(elliptic_output,columns=["h_plus", "h_minus", "h_r","k_r"])
elliptic_df["k_r"] = elliptic_df["k_r"]**2
elliptic_df[["h_plus","k_r"]].to_csv("elliptic_parameters.csv",header=False,index=False)
```
